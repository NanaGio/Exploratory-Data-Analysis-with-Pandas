```python
max = df['ApplicantIncome'].max()
min = df['ApplicantIncome'].min()
intervalo = (max - min) / 5

print(f"Tamanho: {intervalo:.2f}")
```

```python
bucket = [
    min,
    min + intervalo,
    min + 2 * intervalo,
    min + 3 * intervalo,
    min + 4 * intervalo,
    max
]

print(f"Cortes", bucket)
```

```python
import pandas as pd
```

```python
file_path = 'loan_sanction_test.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
female = df[(df['Gender'] == 'Female')]
```

```python
married_females = df[((df['Gender'] == 'Female') & (df['Married'] == 'Yes'))]
```

```python
married_females_Graduate = df[(df['Education'] == 'Graduate') & (df['Gender'] == 'Female') & (df['Married'] == 'Yes')]
media_income = married_females_Graduate['ApplicantIncome'].mean()
print(f"Média: {media_income:.2f}")
```

```python
married_females_NotGraduate = df[((df['Education'] == 'Not Graduate') & (df['Gender'] == 'Female') & (df['Married'] == 'Yes'))]
media_income = married_females_NotGraduate['ApplicantIncome'].mean()
print(f"Média: {media_income:.2f}")
```

```python
married_females_perArea = married_females['Property_Area'].value_counts()
print(married_females_perArea)
```
